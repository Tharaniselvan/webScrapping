```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as ts
```

```python
url='https://www.flipkart.com/search?q=tshirt&as=on&as-show=on&otracker=AS_Query_PredictiveAutoSuggest_6_0_na_na_na&otracker1=AS_Query_PredictiveAutoSuggest_6_0_na_na_na&as-pos=6&as-type=PREDICTIVE&suggestionId=tshirt&requestId=57d7d6e5-2ac0-4190-91bd-2b35640806c3'
```

```python
res = requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
Tshrit_names=soup.select('.IRpwTa')
```

```python
Tshrit_names
```

```python
Tshrit_prices=soup.select('._3bPFwb ._30jeq3')
```

```python
Tshrit_prices
```

```python
Tshrit_offers=soup.select('._3bPFwb span')
```

```python
Tshrit_offers
```

```python
df=ts.DataFrame(Tshrit_names,columns=['Tshrit_names'])
```

```python
df
```

```python
df['Tshrit_prices']=Tshrit_prices
```

```python
Tshrit_prices
```

```python
df['Tshrit_offers']=Tshrit_offers
```

```python
Tshrit_offers
```

```python
df.to_csv('tsp.csv')
```
